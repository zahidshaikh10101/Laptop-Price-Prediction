```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('LaptopPricePrediction.csv')
df.head()
```

```python
df.drop(df.columns[df.columns.str.contains('unnamed', case=False)], axis=1, inplace=True)
df.head()
```

```python
df["Price"] = df["Price"].str.replace(',', '')
df["Price"] = df["Price"].str.replace('₹', '')
df.head()
```

```python
df['RAM'] = df['RAM'].str.replace('\s.*','')
df['Warranty'] = df['Warranty'].str.replace('\s.*','')
df.head()
```

```python
df.drop(['rating', 'Operating System'], axis=1, inplace=True)
df.head()
```

```python
df['Storage'] = df['Storage'].replace('1 TB', '1024 GB', regex=True)
df.head()
```

```python
df['Price'] = df["Price"].astype(str).astype(int)
df.dtypes
```

```python
df.head(10)
```

```python
column = [X for X in df.columns if df[X].dtype == 'O']
column
```

```python
for X in column:
    label= df.groupby([X])['Price'].mean().sort_values().index
    label={k:i for i,k in enumerate(label,0)}
    df[X]=df[X].map(label)
df.head(10)
```

```python
X = df.drop(['Price'], axis=1)
y = df['Price']
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
scaler = sc.fit_transform(X)
X = pd.DataFrame(scaler)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42,shuffle = True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
